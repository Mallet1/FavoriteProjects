```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cancer_classification.csv')
```

```python
df.describe().transpose()
```

```python
sns.countplot(x='benign_0__mal_1',data=df)
# count plots are good for classification to see the number of instances per label and see if it's a well-balanced problem
```

```python
df.corr()['benign_0__mal_1'].sort_values()[:-1].plot(kind='bar')
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr())
```

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(30,activation='relu'))

# sigmoid because this is binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot() # overfitting example (begins increasing)
# use early stopping/callbacks to stop training before overfitting
```

```python
# will use early stopping/callbacks to stop training before overfitting
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(30,activation='relu'))

# sigmoid because this is binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# need a quantity to monitor
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
# mode what you are trying to accomplish with the quantity you are monitoring
# in this case I want to "minimize" val_loss
# patience means we will wait 25 epochs after detecting a stopping point due to noise
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
# next to prevent overfitting is dropout layers
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5)) # rate is the probability that you will randomly turn off a neuron
# 0.5 means that half of the neurons will be turned off randomly during each batch

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

# sigmoid because this is binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot() # it lasted longer which is good because it means it is still learning even on the the later epochs
# lasted longer and didn't overfit due to the earlystop and the dropout layers
```

```python
predictions = (model.predict(X_test) > 0.5).astype("int32") # replacement for model.predict_classes
# model.predict(X_test) > 0.5 returns all booleans then .astype("int32") converts them to ints (1 = True 0 = False)
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
